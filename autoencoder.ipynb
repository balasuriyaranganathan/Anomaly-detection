```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Read the dataset
data = pd.read_csv('111.csv', encoding='latin1')

# Select the relevant features for anomaly detection
features = ['Time', 'Length']

# Extract the feature values from the dataset
X = data[features]

# Normalize the feature values
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Define the autoencoder model
input_dim = X_scaled.shape[1]
encoding_dim = 10  # Adjust the encoding dimension based on your data
autoencoder = Sequential([
    Dense(encoding_dim, activation='relu', input_shape=(input_dim,)),
    Dense(input_dim, activation='sigmoid')
])

# Compile and train the autoencoder model
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=32)

# Use the trained model for anomaly detection
reconstructed = autoencoder.predict(X_scaled)
mse = np.mean(np.power(X_scaled - reconstructed, 2), axis=1)
threshold = np.percentile(mse, 95)  # Adjust the percentile threshold based on your data

# Identify congestion points based on the anomaly scores
congestion_points = data[mse > threshold]
print("Congestion Points:")
print(congestion_points)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
data = pd.read_csv('111.csv', encoding='latin1')
features = ['Protocol', 'Length']
X = data[features]
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
X['Protocol'] = label_encoder.fit_transform(X['Protocol'])
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
input_dim = X_scaled.shape[1]
encoding_dim = 10  # Adjust the encoding dimension based on your data
autoencoder = Sequential([
    Dense(encoding_dim, activation='relu', input_shape=(input_dim,)),
    Dense(input_dim, activation='sigmoid')
])
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=32)
reconstructed = autoencoder.predict(X_scaled)
mse = np.mean(np.power(X_scaled - reconstructed, 2), axis=1)
threshold = np.percentile(mse, 95)  # Adjust the percentile threshold based on your data

# Identify congestion points based on the anomaly scores
Anamaly = data[mse > threshold]
print("Anamaly detected:")
print(Anamaly)
```
